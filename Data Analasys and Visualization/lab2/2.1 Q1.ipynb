```python
#Beautiful Soup


import requests
from bs4 import BeautifulSoup

url = "https://aniwatch.to/top-airing"
url1 = "https://aniwatch.to/one-piece-100"

r = requests.get(url)
s=requests.get(url1)

soup = BeautifulSoup(r.content, "html.parser")
soups = BeautifulSoup(s.content, "html.parser")



names = soup.find_all('a', class_='dynamic-name') # Names of all the top airing shows
description=soups.find('div', class_='text') # Descripton of the best Show
name = soups.find('h2', class_='film-name dynamic-name')  # Name of the best Show


print(name.text)
print(description.text)

print("Top Airing:")
print()



for div in names:
    print(div.text,{div['href']}) # Names and Links of the top airing shows



```
